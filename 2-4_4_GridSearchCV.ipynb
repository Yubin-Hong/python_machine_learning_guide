```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
import pandas as pd

iris = load_iris()
X_train, X_test, Y_train, Y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=121)
dtree = DecisionTreeClassifier()

parameters = {'max_depth': [1, 2, 3],
              'min_samples_split': [2, 3]}

GSDT = GridSearchCV(dtree, param_grid=parameters,
                    scoring='accuracy', cv=3, refit=True)
GSDT.fit(X_train, Y_train)

scores_df = pd.DataFrame(GSDT.cv_results_)
scores_df
```

```python
scores_df[['params', 'mean_test_score', 'rank_test_score', 'split0_test_score', 'split1_test_score', 'split2_test_score']]
```

```python
print('Optimal Parameter by GridSearchCV:', GSDT.best_params_)
print('Highest Accuracy by GridSearchCV:', GSDT.best_score_)
```

```python
estimator = GSDT.best_estimator_

from sklearn.metrics import accuracy_score

pred = estimator.predict(X_test)
accuracy = accuracy_score(pred, Y_test)
print('Test accuracy using best_estimator:', accuracy )
```
