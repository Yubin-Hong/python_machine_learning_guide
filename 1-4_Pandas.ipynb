>Transform list&ndarray into Df  

```python
import numpy as np

#1D data to Dataframe
col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('shape of array1:', array1.shape)

#list -> Df
list1_df = pd.DataFrame(list1, columns = col_name1)
print('Dataframe from list1:\n', list1_df)

#ndarray -> Df
array1_df = pd.DataFrame(array1, columns = col_name1)
print('Dataframe from array1:\n', array1_df)
```

```python
#2D data to Dataframe
col_name2 = ['col1','col2','col3']

list2 = [[1,2,3],[4,5,6]]
array2 = np.array(list2)
print('shape of array2:', array2.shape)

#list -> Df
list2_df = pd.DataFrame(list2, columns = col_name2)
print('Dataframe from list2:\n', list2_df)

#ndarray -> Df
array2_df = pd.DataFrame(array2, columns = col_name2)
print('Dataframe from array2:\n', array2_df)
```

```python
titanic_train_df['Child_or_Adult'] = titanic_train_df['Age'].apply(lambda x: 'Child' if x <= 15 else 'Adult')
titanic_train_df[['Age', 'Child_or_Adult']].head(10)
```

```python
titanic_train_df['Age_Classfication'] = titanic_train_df['Age'].apply(lambda x: 'Child' if x <= 15 else ('Adult' if x>15 and x<=60 else 'Elderly'))
titanic_train_df['Age_Classfication'].value_counts()
```

```python
def age_class(age):

    if age <=5:
        ageclass = 'Baby'
    elif age <=12 and age >5:
        ageclass = 'Child'
    elif age <=18 and age >12:
        ageclass = 'Teeanger'
    elif age <=25 and age >18:
        ageclass = 'Student'
    elif age <=35 and age >25:
        ageclass = 'Young Adult'
    elif age <=60 and age >35:
        ageclass = 'Adult'
    elif age >60:
        ageclass = 'Elderly'
    return ageclass

age_class(45)
```

```python
titanic_train_df['Age_Classfication'] = titanic_train_df['Age'].apply(lambda x: age_class(x))
titanic_train_df.head(20)
```
